```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
direc = 'C:/Users/Bas/Google Drive/Bas Games/Analysis/Loglik/'
surprise = -np.loadtxt(direc + 'loglik_hvh_final_boards.txt')

direc = 'C:/Users/Bas/Google Drive/Bas Games/Analysis/Move statistics/'
m = np.loadtxt(direc + 'move_stats_hvh.txt')
```

```python
for i in range(40):
    ind = np.nonzero(np.logical_and.reduce([m[:,0]==i,m[:,3]>0,m[:,3]<30]))[0]
    plt.plot(surprise[ind-1],np.log(m[ind,5]),'.k',alpha=0.2)
    
plt.xlabel("Log-likelihood of opponent's previous move")
plt.ylabel("Log response time(s)")
plt.show()
```
