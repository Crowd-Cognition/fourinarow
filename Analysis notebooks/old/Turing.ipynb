```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import seaborn as sns
sns.set_style('white')
```

```python
direc='C:/Users/Bas/Google Drive/Bas Games/Analysis/Turing/'
f = open(direc + 'test.txt', 'r')
d=f.readlines()
```

```python
pos=[int(dd.split()[0]) for dd in d]
correct=[int(dd.split()[1]) for dd in d]
ncorr=np.bincount(pos,weights=correct).astype(int)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 6))

axes.hist(ncorr, normed=False, color='grey', edgecolor='white', bins=np.arange(0, 18, 1))
x = np.arange(0, 18, 1)
y = 180*sts.binom.pmf(x, 18, 0.55)
axes.plot(x, y, color='black')
plt.setp(axes, xlabel='Number of subjects that correctly categorized video', ylabel='Number of videos')
axes.legend()
sns.despine()
plt.show()
```
