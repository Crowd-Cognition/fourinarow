```python
import numpy as np
import matplotlib.pyplot as plt
import pylab 
from scipy import stats
import math
import matplotlib as mpl
from matplotlib import rc,rcParams

fig_direc = 'C:/Users/Bas/Google Drive/Bas Zahy Gianni - Games/Manuscripts/Paper/Figures/'
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] #for \text command
rcParams['font.sans-serif'] = "Helvetica Neue LT Std"
rcParams['font.family'] = "sans-serif"

direc = 'C:/Users/Bas/Google Drive/Bas Games/'
```

```python
lesions = ['final_noprune','final_notree','final_nodelta','final_noact','final_nocenter',
          'final_no2conn','final_no2unc','final_no3','final_no4']
j = [0,1,2,4,5,6,7,8,9]

param_names = [r'$\theta$',r'$\gamma$',r'$\delta$',r'$\lambda$',r'$C$',r'$w_{\text{center}$',
               r'$w_{\text{connected 2-in-a-row}}$',r'$w_{\text{unconnected 2-in-a-row}}$',
              r'$w_{\text{3-in-a-row}}$',r'$w_{\text{4-in-a-row}}$']

Nparams = len(param_names)

def ave_across_splits(p):
    return np.nanmean(np.reshape(p,[int(p.shape[0]/5),5,int(p.shape[1])]),axis=1)

def shrink_input(p):
    return np.column_stack([p[:,[1,2,41,3]],np.nanmean(p[:,24:28]/p[:,7:11],axis=1),p[:,6:11]])
```

```python
Pfull = ave_across_splits(shrink_input(np.loadtxt(direc + 'Analysis/Params/params_hvh_final.txt')))

Plesions=[ave_across_splits(shrink_input(np.loadtxt(direc + 'Analysis/Params/params_hvh_' + m + '.txt')))
             for m in lesions]
```

```python
c1 = np.vstack([[stats.spearmanr(Pfull[:,j[k]]-Plesions[i][:,j[k]],Pfull[:,j[i]])[0] 
                for i in range(9)] for k in range(9)])
p1 = np.vstack([[stats.spearmanr(Pfull[:,j[k]]-Plesions[i][:,j[k]],Pfull[:,j[i]])[1] 
                for i in range(9)] for k in range(9)])

c2 = np.vstack([[stats.spearmanr(Plesions[i][:,j[k]],Pfull[:,j[k]])[0] 
                for i in range(9)] for k in range(9)])
p2 = np.vstack([[stats.spearmanr(Plesions[i][:,j[k]],Pfull[:,j[k]])[1] 
                for i in range(9)] for k in range(9)])

D = np.vstack([[stats.ks_2samp(Plesions[i][:,j[k]],Pfull[:,j[k]])[0]
                for i in range(9)] for k in range(9)])
p3 = np.vstack([[stats.ks_2samp(Plesions[i][:,j[k]],Pfull[:,j[k]])[1]
                for i in range(9)] for k in range(9)])
```

```python
def get_alpha_fdr(alpha,allps):
    return allps[np.nonzero(allps > np.arange(1,len(allps)+1)/len(allps)*alpha)[0][0]]

def plot_with_stars(r,p,colormap,limits,filename,letter):
    fig,ax = plt.subplots()
    
    nstars = (p < get_alpha_fdr(0.05,np.sort(p.flatten()))).astype(int) + \
             (p < get_alpha_fdr(0.01,np.sort(p.flatten()))).astype(int) + \
             (p < get_alpha_fdr(0.001,np.sort(p.flatten()))).astype(int)
        
    for i in range(r.shape[0]):
        for k in range(r.shape[1]):
            if(nstars[i][k]>0):
                ax.text(k,i,'*'*nstars[i][k],ha='center',va='center')
    
    ax.set_xticks(range(Nparams))
    ax.set_yticks(range(Nparams))
    ax.set_xticklabels(param_names,usetex=True,rotation=60,ha='right',fontsize = 14)
    ax.set_yticklabels(param_names,usetex=True,fontsize = 14)    
    ax.text(0.05, 0.9, letter, horizontalalignment='center',
            verticalalignment='center', transform=fig.transFigure, fontsize = 20)
    cax = ax.imshow(r,interpolation='nearest',cmap=colormap,vmin=limits[0],vmax=limits[1])
    cbar = fig.colorbar(cax, ticks=[-1, 0, 1], orientation='vertical')
    fig.tight_layout()
    fig.savefig(fig_direc + filename)
    plt.show()
```

```python
plot_with_stars(c1,p1,'bwr',[-1,1],'tradeoffs_B.pdf','B')
plot_with_stars(D,p3,'Greens',[0,1],'tradeoffs_A.pdf','A')
```

```python
def rosenbrock(x,y):
    return (1-x)**2 + 100*(y-x**2)**2

x = np.arange(-3.0,3.0,0.005)
y = np.arange(-2.0,5.0,0.005)

X,Y = np.meshgrid(x, y)
Z =rosenbrock(X,Y)
plt.pcolormesh(X,Y,Z,cmap = 'inferno')
plt.show()
```
