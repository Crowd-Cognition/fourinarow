```python
def correlate_rt(tpred,tobs,ind,ctype):
    if sum(ind)<=1:
        return np.nan
    if ctype == "spearman":
        return stats.spearmanr(tpred[ind],tobs[ind])[0]
    elif ctype == "pearson":
        return stats.pearsonr(np.log(tpred[ind]+0.5),np.log(tobs[ind]))[0]
    else:
        print("invalid correlation type")
        return np.nan

def correlate_rt_pred_obs(a,b,tpred,ctype):
    temp = np.empty(shape=tpred.shape[0]) * np.nan
    for i in range(40):
        ind=np.logical_and.reduce([t[:,0]==i,n>=a,n<b])
        if sum(ind)>1:
            slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(tpred[ind]+0.5),np.log(t[ind,2]))
            temp[ind] = np.exp(np.log(tpred[ind]+0.5)*slope+intercept)
        elif sum(ind) == 1:
            temp[ind] = t[ind,2]
    ind=np.logical_and(n>=a,n<b)
    return correlate_rt(temp,t[:,2],ind,ctype)

def compute_correlation_subject(a,b,i,tpred,ctype):
    ind=np.logical_and.reduce([t[:,0]==i,n>=a,n<b])
    return correlate_rt(tpred,t[:,2],ind,ctype)
```

```python
C_pred_obs_A = [[correlate_rt_pred_obs(a,30,tpred,ctype) for a in range(15)]
    for ctype,tpred in zip(["spearman","spearman","pearson","pearson"],[t[:,3],thick,t[:,3],thick])]
C_pred_obs_B = [[correlate_rt_pred_obs(5,b,tpred,ctype) for b in range(20,36)]
    for ctype,tpred in zip(["spearman","spearman","pearson","pearson"],[t[:,3],thick,t[:,3],thick])]
C_pred_obs_C = [[np.nanmean([compute_correlation_subject(a,30,i,tpred,ctype) 
                             for i in range(40)]) for a in range(15)]
                for ctype,tpred in zip(["spearman","spearman","pearson","pearson"],
                                       [t[:,3],thick,t[:,3],thick])]
C_pred_obs_D = [[np.nanmean([compute_correlation_subject(5,b,i,tpred,ctype) 
                                          for i in range(40)]) for b in range(20,36)]
                for ctype,tpred in zip(["spearman","spearman","pearson","pearson"],
                                       [t[:,3],thick,t[:,3],thick])]
```

```python
fig,ax=plt.subplots()
ax.set_xlabel('Opening cutoff')
ax.set_ylabel('Correlation coefficient')
ax.axvline(x=5,linestyle='dashed',color='black')
for c,label in zip(C_pred_obs_A,['Spearman tree size', 'Spearman Hick-Hyman',
                           'Pearson tree size', 'Pearson Hick-Hyman']):
    ax.plot(c,label=label)
ax.legend(frameon=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.text(0.02, 0.9, 'A', horizontalalignment='center',
        verticalalignment='center', transform=fig.transFigure, fontsize = 20)
fig.tight_layout()
fig.savefig(fig_direc + 'rt_robustness_A.pdf')
plt.show()
```
