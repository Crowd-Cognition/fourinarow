```python
def plot_histogram(x,Nbins,xname,xlimits,filename,letter):
    fig,ax = plt.subplots()
    ax.hist(x,bins = Nbins)
    ax.set_xlabel(xname)
    ax.set_xlim(xlimits)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.text(0.02, 0.9, letter, horizontalalignment='center',
        verticalalignment='center', transform=fig.transFigure, fontsize = 20)
    ax.set_ylabel('Frequency')
    ax.set_yticks([])
    fig.tight_layout()
    fig.savefig(fig_direc + filename)
    plt.show()
```

```python
import statsmodels.api as sm
import sklearn.linear_model as linear_model

def get_ranks(x):
    temp = np.argsort(x)
    ranks = np.empty(len(x), int)
    ranks[temp] = np.arange(len(x))
    return ranks

def fit_ols(X,y,alpha=0.05):
    if X.shape[0]!=len(y):
        X=X.T
    X = sm.add_constant(X,prepend=False,has_constant='skip')
    model = sm.OLS(y, X)
    res = model.fit()
    return res.params,res.conf_int(alpha)

def fit_ard(X,y):
    if X.shape[0]!=len(y):
        X=X.T
    model = linear_model.ARDRegression()
    model.fit(X,y)
    return model.coef_

def fit_lasso(X,y):
    if X.shape[0]!=len(y):
        X=X.T
    model = linear_model.Lasso()
    model.fit(X,y)
    return model.coef_

def fit_ridge(X,y):
    if X.shape[0]!=len(y):
        X=X.T
    model = linear_model.Ridge()
    model.fit(X,y)
    return model.coef_
```

```python
P=np.loadtxt(direc + 'Analysis/Params/params_hvh_final_short.txt')

X = np.vstack([get_ranks(P[:N,i]) for i in range(10)])
y = get_ranks(np.mean(computational_time,axis=1))

plt.plot(np.corrcoef(X,y)[-1][:-1])

params,conf = fit_ols(X,y,0.05)
plt.errorbar(range(len(params)-1),params[:-1], 
             yerr=[params[:-1]-conf[:-1,0], conf[:-1,1]-params[:-1]])

params = fit_ard(X,y)
plt.plot(range(len(params)),params)

params = fit_lasso(X,y)
plt.plot(range(len(params)),params)

params = fit_ridge(X,y)
plt.plot(range(len(params)),params)

plt.show()
```
